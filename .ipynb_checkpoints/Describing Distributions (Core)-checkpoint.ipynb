```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
from scipy import stats
```

# Load Data

```python
df = pd.read_csv('data/medical_data.csv')
df.head()
```

```python
df.info()
```

# Features to Analyze

### VitD_levels

```python
# Plot a histogram with a kde (kernel density estimate)

# Plot and standardize for use with other variables
col = 'VitD_levels'
feature = df[col]
sns.kdeplot(feature);
```

```python
# Add a line for the mean (red)
# Add a line for the median (green)

# Assign values for mean and median
mean = feature.mean()
median = feature.median()


# Replot with lines
fig,ax = plt.subplots(figsize=(12,8))
sns.kdeplot(feature)
ax.axvline(mean, color= 'red')
ax.axvline(median, color='green');
```

```python
# Add a line for for +1 std from the mean (black)
# Add a line for the - 1 std from the mean (black)
# Highlight the range between +1 and =1 std (yellow)

# Assign values for one std above and below mean
std = feature.std()
plus_one_std = mean + std
minus_one_std = mean - std
```

```python
# Replot with better labeling
col = 'VitD_levels'
feature = df[col]
mean = feature.mean()
median = feature.median()
std = feature.std()
plus_one_std = mean + std
minus_one_std = mean - std
fig,ax = plt.subplots(figsize=(12,8))
sns.kdeplot(feature, ax=ax)
ax.axvline(mean, color= 'red', label=f'mean = {mean:,.2f}')
ax.axvline(median, color='green', label=f'median = {median:,.2f}')
# Lines for std's
ax.axvline(plus_one_std, color = 'black',label=f'+1 std = {plus_one_std:,.2f}')
ax.axvline(minus_one_std, color = 'black', label = f'-1 std = {minus_one_std:,.2f}')
# highlight range betweem std's
ax.axvspan(plus_one_std, minus_one_std, color = 'yellow', zorder = 0)
ax.set_title(f'{col}')
ax.legend();
```

### 2. Does it have a skew? If so, which direction (+/-)

VitD_levels feature has nearly symmetrical mean and median. Due to this, it would be considered to be a normal curve.

Doc_visits feature has nearly symmetrical mean and median. Due to this, it would be considered to be a normal curve.

TotalCharge feature seems to display a positive skew for the following reasons.
- The median is smaller than the mean
- The right tail end of the distribution is longer in comparison to the left

### 3. What type of kurtosis does it display? (Mesokurtic, Leptokurtic, Platykurtic)

Using the Pearson calculation:

- Mesokurtic (Kurtosis ~3)

- Leptokurtic (Kurtosis >3)     

- Platykurtic (Kurtosis < 3) 

Mesokurtic (values cluster in the middle of the distribution - normal)

Leptokurtic ("pointy", values cluster in the peak of the distribution - positive)

Platykurtic ("flat", values cluster in the tails of the distribution - negative)

```python
# VitD_levels

kurt_VitD = stats.kurtosis(df['VitD_levels'], fisher = False)
kurt_VitD
```

VitD_levels displays Mesokurtic

```python
# Doc_visits

kurt_Doc = stats.kurtosis(df['Doc_visits'], fisher = False)
kurt_Doc
```

Doc_visits displays Mesokurtic

```python
# TotalCharge

kurt_Total = stats.kurtosis(df['TotalCharge'], fisher = False)
kurt_Total
```

TotalCharge displays Leptokurtic
- Leptokurtic refers to the "pointy" feature the curve would present
- Values would cluster in the peak of the distribution
